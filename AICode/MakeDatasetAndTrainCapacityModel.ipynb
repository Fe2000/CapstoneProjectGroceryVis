# Data and Preprocessing

```python
path = 'dataFinalSubBinary.txt'
# path = 'data.txt'
```

```python
data = pd.read_csv(path)
```

```python
data
```

```python
images = [ np.asarray((Image.open(x)), dtype = 'float32')/255  for x in data['Files'] ]
```

```python
Y = data.iloc[ : , 1: ]
Y
```

```python

enc = OneHotEncoder()

Y = enc.fit_transform(Y).toarray()

Y = pd.DataFrame(Y)
```

```python
Y
```

```python
images = np.asarray(images)
images = images.reshape(  len(images), 300, 420, 3 )
```

```python
x_train , x_test , y_train , y_test = train_test_split(images,Y, test_size=0.2)
```

# Model

```python
model = Sequential()

model.add( Input(shape=(300,420 ,3), dtype="float32"   )  )
model.add(Conv2D(48, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(24, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(24, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(16, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(16, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(8, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add( Dense(2, activation='sigmoid')  )


model.summary()
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy' , 'Recall' , 'Precision'])
```

# Training

```python
model.fit( x_train, y_train, epochs=10, batch_size=4, validation_split=0.1)
```

# Evaluation

```python
model.evaluate(x_test,y_test)
```

# Prediction

```python
def get_pred(arg1):
    modelTest = keras.models.load_model('CapacityDetectionV7')

    im = Image.open(arg1)

    imResize = im.resize((300,420), Image.ANTIALIAS)
    imResize.save("TestSpot/TempPic"+'.jpg', 'JPEG', quality=90)

    image1 = np.asarray((Image.open("TestSpot/TempPic.jpg")), dtype = 'float32')/255
    
    image1 = np.asarray(image1)
    image1 = image1.reshape(  1, 300, 420, 3 )        

    result = modelTest.predict(image1)
        
    tClass = result.max()
    
    if(tClass == result[0][0]):
        
        print('Empty')

    if(tClass == result[0][1]):
        
        print('Full')
        
    if(tClass == result[0][2]):
        
        print('High')    

    if(tClass == result[0][3]):
        
        print('Low')
        
    if(tClass == result[0][4]):
        
        print('Middle')   
        
    print(result)
```
